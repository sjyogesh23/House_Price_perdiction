```python
import pandas as pd
```

```python
df1 = pd.read_csv("train.csv")
df2 = pd.read_csv("test.csv")
```

```python
threshold = len(df1) * 0.5
df1 = df1.dropna(axis=1, thresh=threshold)
```

```python
train_X = df1.drop(["SalePrice", "FireplaceQu", "Id"], axis=1)
train_X.head()
```

```python
train_y = df1.SalePrice
train_y.head()
```

```python
threshold = len(df2) * 0.5
df2 = df2.dropna(axis=1, thresh=threshold)
```

```python
val_X = df2
val_X = df2.drop("Id", axis=1)
val_X.head()
```

```python
print(len(val_X.columns))
print(len(train_X.columns))
train_X.columns.difference(val_X.columns)
```

```python
from sklearn.preprocessing import OneHotEncoder

categorical_cols = []
for col in train_X.columns:
    if train_X[col].dtype == 'object':
        categorical_cols.append(col)

print("Categorical columns:", categorical_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

for col in categorical_cols:
    train_X[col] = label_encoder.fit_transform(train_X[col])
    val_X[col] = label_encoder.fit_transform(val_X[col])

val_X.head()
```

```python
train_X.fillna(train_X.mean(), inplace=True)
val_X.fillna(train_X.mean(), inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor()
model.fit(train_X,train_y)
```

```python
final_model = RandomForestRegressor(random_state=1)
final_model.fit(train_X, train_y)
```

```python
from sklearn.metrics import mean_absolute_error
y_pred = final_model.predict(train_X)
mean_absolute_error(train_y, y_pred)
```

```python
from sklearn.metrics import r2_score

r_squared = r2_score(train_y, y_pred)
print("R-squared:", r_squared)
```

```python
val_y = final_model.predict(val_X)
val_y
```

```python
df = pd.DataFrame(val_y, columns=['SalePrice'], index=range(1461, 1461+len(val_y)))
df.index.name = 'Id'
print(df)
```

```python
df.to_csv("Home_Ans_v2.csv")
```
